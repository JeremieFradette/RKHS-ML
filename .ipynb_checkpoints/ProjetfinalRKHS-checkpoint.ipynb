```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles, make_moons
from sklearn.decomposition import PCA, KernelPCA
from sklearn.preprocessing import StandardScaler
```

```python
# 1. Créer deux jeux de données non linéaires
# Cercles concentriques (bruit faible)
X_circles, y_circles = make_circles(n_samples=300, factor=0.5, noise=0.05, random_state=42)
# Lune (two moons)
X_moons, y_moons = make_moons(n_samples=300, noise=0.05, random_state=42)

# Standardiser (important pour PCA et kernel PCA)
scaler = StandardScaler()
X_circles_scaled = scaler.fit_transform(X_circles)
X_moons_scaled = scaler.fit_transform(X_moons)
```

```python
# 2. Appliquer PCA linéaire (2 composantes)
pca_lin = PCA(n_components=2)
X_circles_pca = pca_lin.fit_transform(X_circles_scaled)
X_moons_pca = pca_lin.fit_transform(X_moons_scaled)
```

```python
# 3. Appliquer Kernel PCA avec noyau RBF (gamma à ajuster)
kpca_rbf = KernelPCA(n_components=2, kernel='rbf', gamma=1.0)  # gamma élevé pour cercles
X_circles_kpca = kpca_rbf.fit_transform(X_circles_scaled)

kpca_rbf2 = KernelPCA(n_components=2, kernel='rbf', gamma=0.5)  # gamma un peu plus faible pour moons
X_moons_kpca = kpca_rbf2.fit_transform(X_moons_scaled)
```

```python
# 4. Visualisation
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

# Ligne 1 : Cercles concentriques
# Données originales
axes[0, 0].scatter(X_circles[:, 0], X_circles[:, 1], c=y_circles, cmap='bwr', edgecolors='k', s=40)
axes[0, 0].set_title("Cercles (original)")
axes[0, 0].set_aspect('equal')

# PCA linéaire
axes[0, 1].scatter(X_circles_pca[:, 0], X_circles_pca[:, 1], c=y_circles, cmap='bwr', edgecolors='k', s=40)
axes[0, 1].set_title("PCA linéaire")
axes[0, 1].set_aspect('equal')

# Kernel PCA
axes[0, 2].scatter(X_circles_kpca[:, 0], X_circles_kpca[:, 1], c=y_circles, cmap='bwr', edgecolors='k', s=40)
axes[0, 2].set_title("Kernel PCA (RBF, gamma=1.0)")
axes[0, 2].set_aspect('equal')

# Ligne 2 : Lune
# Original
axes[1, 0].scatter(X_moons[:, 0], X_moons[:, 1], c=y_moons, cmap='bwr', edgecolors='k', s=40)
axes[1, 0].set_title("Lune (original)")
axes[1, 0].set_aspect('equal')

# PCA linéaire
axes[1, 1].scatter(X_moons_pca[:, 0], X_moons_pca[:, 1], c=y_moons, cmap='bwr', edgecolors='k', s=40)
axes[1, 1].set_title("PCA linéaire")
axes[1, 1].set_aspect('equal')

# Kernel PCA
axes[1, 2].scatter(X_moons_kpca[:, 0], X_moons_kpca[:, 1], c=y_moons, cmap='bwr', edgecolors='k', s=40)
axes[1, 2].set_title("Kernel PCA (RBF, gamma=0.5)")
axes[1, 2].set_aspect('equal')

plt.tight_layout()
plt.show()
```
